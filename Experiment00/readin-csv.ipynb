```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Specify column names
column_names = ["Frame NO.", "Target Total", "Target NO.", "X(m)", "Y(m)", "Z(m)", "Velocity(m/s)", "Intensity", "Intensity(Log)", "Extra"]

# Read CSV data, skip the first row, and consider the trailing comma
df = pd.read_csv('slope-diff-data\\1\\xwr16xx_processed_stream_2023_09_14T16_41_23_134.dat', header=None, names=column_names, skiprows=1)

# Drop the extra column created due to the trailing comma
df = df.drop(columns=["Extra"])

# Strip leading/trailing whitespaces from column names
df.columns = df.columns.str.strip()
```

```python

# Create a new figure
fig = plt.figure()

# Create a three-dimensional axes
ax = fig.add_subplot(111, projection='3d')

# Plot each point
for index, row in df.iterrows():
    ax.scatter(row['X(m)'], row['Y(m)'], row['Z(m)'])

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show plot
plt.show()
```
